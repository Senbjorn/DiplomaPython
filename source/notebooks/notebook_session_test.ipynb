```python
with open(str(project.output_path / "benchmark_02" / "sessions" / "benchmark_02.session"), "rb") as input_file:
    session = pickle.load(input_file)
```

```python
for model_name, model in session["models"].items():
    print(model_name)
    print(model["rmsd_record"]["native-start"])
    print(model["rmsd_record"]["native-result"])
    print(model["rmsd_record_rigid"]["native-result"])
    print(model["native_energy"])
    print(model["result"]["energies"][-1])
    print(model["result"]["energies"][0])
    print(model["angle"])
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
ax.plot(range(0, 5), range(0, 5), marker='s', markersize=9, linewidth=1)
ax.set_xlabel("Value, $\\log_{10}\\left(\\frac{kJ}{mol}\\right)$")
plt.show()
```

```python
for model_name, model in session["models"].items():
    print(model_name)
    print("\tinitial energy:               ", int(model["native_energy"]))
    print("\tbefor optimization:           ", int(model["result"]["energies"][0]))
    print("\tafter optimization (flexible):", int(model["result"]["energies"][-1]))
    print("\tinitial rmsd:                 ", f'{model["rmsd_record"]["native-start"]["total"]:.2f}')
    print("\tfinal rmsd (flexible):        ", f'{model["rmsd_record"]["native-result"]["total"]:.2f}')
    print("\tfinal rmsd (rigid):           ", f'{model["rmsd_record_rigid"]["native-result"]["total"]:.2f}')
    print()
```
