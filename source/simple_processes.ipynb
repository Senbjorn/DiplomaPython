```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import time
from matplotlib import style

from multiprocessing import Queue, Process, Lock
from queue import Empty

style.use("fivethirtyeight")
```

```python
def update_data(q_x, q_y):
    xlist = []
    ylist = []
    while True:
        time.sleep(0.05)
        x, y = np.random.normal(0, 1, 2)
        q_x.put(x)
        q_y.put(y)

%matplotlib
fig, ax = plt.subplots() 

x = []
y = []
x_queue = Queue()
y_queue = Queue()

def animate(i):
    try:
        x.append(x_queue.get(False))
        y.append(y_queue.get(False))
    except Empty:
        pass
    ax.clear()
    ax.scatter(x, y, color="red")

uprocess = Process(target=update_data, args=(x_queue, y_queue), daemon=True)
uprocess.start()
    
anim = animation.FuncAnimation(fig, animate, 30)
```

```python
def object_transfer(obj_q, obj_lock):
    x = [8, 5, 1]
    y = [4, 2, 1]
    d = {"x": x, "y": y}
    print("on lock")
    with obj_lock:
        obj_q.put(d)
    print("on sleep")
    time.sleep(2)
    with obj_lock:
        print(d)

obj_lock = Lock()
obj_q = Queue()

uprocess = Process(target=object_transfer, args=(obj_q, obj_lock))
uprocess.start()
time.sleep(2)
with obj_lock:
    d = obj_q.get()
    d["x"][0] = 0
    print(d)
```

```python
uprocess.terminate()
```
