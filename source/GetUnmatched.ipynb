```python
# openmm
molecule = app.PDBFile("../data/6awr.pdb")
forcefield = app.ForceField("charmm36.xml")
modeller = app.Modeller(molecule.topology, molecule.positions)
# prody
pdb_prot = pdy.parsePDB("../data/6awr.pdb")
psf_prot = pdy.parsePSF("/home/semyon/6awr_autopsf.psf")
```

```python
# get one of wrong residues (index 26)
unmatched = None
chain_gen = modeller.topology.chains()
for chain in chain_gen:
    residue_gen = chain.residues()
    for residue in residue_gen:
        print(residue)
        if residue.index == 26:
            unmatched = residue
            break
print(unmatched)
```

```python
# print residue bonds
count = sum([1 for atom in unmatched.atoms()])
for atom in unmatched.atoms():
    print(atom)
print(count)
```
